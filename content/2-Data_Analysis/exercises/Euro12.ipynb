### Step 14. Present only the Shooting Accuracy from England, Italy and Russia

```python
import pandas as pd

url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
df = pd.read_csv(url)

# Clean percent sign and convert to numeric
df["Shooting Accuracy"] = (
    df["Shooting Accuracy"].astype(str)
      .str.replace("%", "", regex=False)
      .str.strip()
)
df["Shooting Accuracy"] = pd.to_numeric(df["Shooting Accuracy"], errors="coerce")

teams = ["England", "Italy", "Russia"]
result = (
    df[df["Team"].isin(teams)][["Team", "Shooting Accuracy"]]
      .set_index("Team")
      .loc[teams]
      .reset_index()
)

# Format as percentages for display
result["Shooting Accuracy"] = result["Shooting Accuracy"].map(lambda v: f"{v:.1f}%" if pd.notna(v) else "")
print(result)
```
